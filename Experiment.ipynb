# Кратчайшие пути в графе: алгоритм Дейкстры на метках

```python
from my_graph import Graph, MyNx, experiment
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%%time
g = Graph(n=10,
          m=90,
          q=10,
          r=10**6,
          directed_graph = True,
          random_seed=321121)
mynx = MyNx(g)
TETA, label,time_ = mynx.dijkstra_predecessor_and_distance(source=8)
```

```python
N= range(10,10**3+1,10)
M = [round(n**2/10) for n in N]
q=[1]*len(M)
r=[10**6]*len(M)
res1 = experiment(N=N, M=M, q=q, r=r,m_label = 'm=n^2/10', source = 8)
M = [round(n**2-n) for n in N]
res2 = experiment(N=N, M=M, q=q, r=r,m_label = 'm=n^2', source = 8)
res1 = pd.concat([res1,res2],axis=0)
```

```python
plt.figure(figsize=(11,6.5))
ax = sns.lineplot(x = res1['n'], y = res1['Time, sec'],hue= res1['label'])
ax.axes.set_title("Временная сложность алгоритма Дейкстри на основе метках",fontsize=16)
ax.set_xlabel("Кол-во вершин",fontsize=14)
ax.set_ylabel("Time, sec",fontsize=14)
plt.grid()
```

```python
N= range(101,10**3+1,10)
M = [10*n for n in N]
q=[1]*len(M)
r=[10**6]*len(M)
res1 = experiment(N=N, M=M, q=q, r=r,m_label = 'm=10*n', source = 8)
M = [round(100*n) for n in N]
res2 = experiment(N=N, M=M, q=q, r=r,m_label = 'm=100*n', source = 8)
res2 = pd.concat([res1,res2],axis=0)
```

```python
plt.figure(figsize=(11,6.5))
ax = sns.lineplot(x = res2['n'], y = res2['Time, sec'],hue= res2['label'])
ax.axes.set_title("Временная сложность алгоритма Дейкстри на основе метках",fontsize=16)
ax.set_xlabel("Кол-во вершин",fontsize=14)
ax.set_ylabel("Time, sec",fontsize=14)
plt.grid()
```
